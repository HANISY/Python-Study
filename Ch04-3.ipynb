```python
# for 반복문과 범위 조합
for i in range(5):
    print(str(i) + "=반복변수")
print()

for i in range(5,10):
    print(str(i) + "=반복변수")
print()

for i in range(0,10,3):
    print(str(i) + "=반복변수")
print()
```

```python
array = [273, 32, 103, 57, 52]

for i in range(len(array)):
    print("{}번째 반복: {}".format(i, array[i]))
```

```python
# 역반복문

for i in range(4,0 - 1,-1):
    print("현재 반복 변수: {}".format(i))

print()

for i in reversed(range(5)):
    print("현재 반복 변수: {}".format(i))
```

```python
# 반복문으로 피라미드 만들기 (1)

output = ""

for i in range(1, 5): 
    for j in range(0, i):
        output += "*" # 내부 반복문은 * 출력
    output += "\n" # 외부 반복문은 줄바꿈 

print(output)
```

```python
# 반복문으로 피라미드 만들기 (2)

output = ""

for i in range(1, 5):
    for j in range(9, i, -1):
        output += ' ' # 공백 생성
    for k in range(0, 2 * i -1):
        output += '*' # 별 생성
    output += '\n' # 줄 바꿈 

print(output)
```

```python
# 문자열 * 연산자로 피라미드 만들기 
output = ""

for i in range(1, 5):
    output += ("*" * i)
    output += "\n"

print(output)
```

```python
# while 반복문 : for 반복문처럼 사용하기 
i = 0 
while i < 5:
    print("{}번째 반복입니다.".format(i))
    i += 1
```

```python
list_test = [1, 2, 1, 2]
value = 2

while value in list_test:
    list_test.remove(value) # 리스트 내부의 모든 2가 제거될 때까지 반복

print(list_test)
```

```python
import time
```

```python
number = 0

target_tick = time.time() + 5
while time.time() < target_tick:
    number += 1

print("5초 동안 {}번 반복했습니다.".format(number))
```

```python
# 빈 칸을 채워 키와 값으로 이루어진 각 리스트를 조합해 하나의 딕셔너리 만들기

key_list = ["name", "hp", "mp", "level"]
value_list = ["기사", 200, 30, 5]
character = {}

for i in range(0, len(key_list)):
    character[key_list[i]] = value_list[i]

print(character)
```

```python
# 1부터 시작해 숫자를 하나씩 더해 나갈 때, 합이 10,000을 처음으로 넘기는 마지막 숫자 구하기

limit = 10000
i = 1

sum_value = 0 
while sum_value < limit:
    sum_value += i 
    i += 1

print("{}를 더할 때 {}를 넘으며 그때의 값은 {}입니다.".format((i-1), limit, sum_value)) # 괄호 잘 구분하기 
```

```python
# 1부터 100까지의 두 수를 더해 100이 되도록 짝지어 곱했을 때, 최댓값이 되는 경우 찾기

max_value = 0
a = 0 
b = 0

for i in range(1,100):
    j = 100 - i 
    current = i * j 
    if max_value < current:
        a = i 
        b = j  
        max_value = current # 순서 반대로 작성하면 답이 다르게 나오니 주의!

print("최대가 되는 경우: {} * {} = {}".format(a, b, max_value))
```
