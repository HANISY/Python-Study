##### 초급 문제 - 문자열 대소문자 변환기
- **문제 설명**: <br>
문자열을 입력받아, 소문자는 대문자로, 대문자는 소문자로 바꿔 출력하시오.<br>
공백이나 숫자는 그대로 둔다.

- **입력 예시**: <br>
```PyThOn```

- **출력 예시**: <br>
```pYtHoN```

- **핵심 개념**: <br>
문자열 순회, isupper(), islower(), upper(), lower() 활용

````python
text = input("문자열 입력> ")

result = ''     # 바꾼 문자열을 저장할 공간 

for char in text:
    if char.isupper():     # isuper()는 문자열 객체의 메서드기 때문에 객체 뒤에 점(.)을 붙여야 함 
        result += char.lower()     # 새 문자열을 반환하지만, 원본 문자는 변하지 않으므로 바꾼 값을 '누적해서' 저장할 공간 필요 
    else:
        result += char.upper()

print(result)
````

````python
# chatGPT가 단순화해 준 답안

text = input("문자열 입력> ")

result = ''.join(
    char.lower() if char.isupper() else char.upper()
    for char in text
)

print(result)
````

##### 중급 문제 - 단어별 뒤집기
- **문제 설명**: <br>
문장에서 각 단어의 철자 순서를 뒤집되, 단어의 순서는 그대로 유지하시오.<br>
공백이나 숫자는 그대로 둔다.

- **입력 예시**: <br>
```python is fun```

- **출력 예시**: <br>
```nohtyp si nuf```

- **핵심 개념**: <br>
```split()```, 슬라이싱 ```[::-1]```,```''.join()``` 활용

````python
text = input("문자열 입력> ")

words = text.split()

reversed_words = words[::-1]

result = ' '.join(reversed_words)

print(result)
````

````python
text = input("문자열 입력> ")

words = text.split()        # 문장을 공백 기준으로 단어 리스트로 분리

reversed_words = []     # 반복문에서 결과를 저장하기 위한 공간(리스트) 지정

for word in words:
    rev_word = word[::-1]
    reversed_words.append(rev_word)

result = ' '.join(reversed_words)       # 공백을 구분자로 사용

print(result)
````

##### 고급 문제 - 압축 문자열 (Run-Length Encoding)
- **문제 설명**: <br>
문자열이 주어졌을 때, 같은 문자가 연속으로 반복되면<br>
반복 횟수를 숫자로 표시해 압축하시오. <br>
(예: ```aaabbc```→ ```a3b2c1```)

- **입력 예시 1**: <br>
```aaabbc```<br>
→ 출력: ```a3b2c1```

- **입력 예시 2**: <br>
```aabbaa```<br>
→ 출력: ```a2b2a2```

- **핵심 개념**: <br>
반복문, 이전 문자 비교, 누적 카운트, 문자열 누적

````python
text = input("문자열 입력> ")

result = ''

previous = text[0]      # 첫 글자를 ‘현재 연속 중인 문자’로 기준 설정
count = 1

for i in range(1,len(text)):        # 두 번째 글자(index 1)부터 비교 시작, 문자열 길이 기준으로 반복 
    current = text[i]

    if current == previous: 
        count += 1 
    else: 
        result = result + previous + str(count)     # previous는 문자열, count는 정수 → 문자열 결합을 위해 str()로 변환 필요
        previous = current      # 다음 연속 구간을 새 문자(current)로 갱신
        count = 1

result = result + previous + str(count)     # 반복문 안에서는 "문자가 바뀔 때"만 저장되므로 마지막 문자 구간은 별도로 저장해야 함 

print(result)
````
