#### 문제 1. 학생 점수 통계 계산 함수

##### ✓ 요구사항 
여러 학생의 점수를 입력받아, 
- 최고 점수
- 최저 점수
- 평균 점수
- **합격자 수(60점 이상)**

를 계산하는 프로그램을 작성하시오

1. 점수는 첫 줄에 `"70 80 90 100"`처럼 **공백으로 구분된 정수**로 입력된다

2. 결과는 아래 형식으로 출력한다 

```text
scores: 70 80 90 100
max: 100
min: 70
avg: 85.0
pass: 4

````python
scores = list(map(int, input("점수 입력: ").split()))

print(f"scores: {' '.join(map(str, scores))}")
print(f"max: {max(scores)}")
print(f"min: {min(scores)}")
print(f"avg: {sum(scores)/len(scores)}")

pass_cnt = 0 
for score in scores:
    if score >= 60:
        pass_cnt += 1 
print(f"pass: {pass_cnt}")
````

```python
# 방법 1: 불리언(True=1, False=0)을 활용

# 방법 2: 리스트 컴프리헨션을 사용한 필터링 (합격 점수만 새 리스트로 만든 후 개수 세기)
pass_cnt = len([score for score in scores if score >= 60])

#### 문제 2. 최소 길이 이상인 단어만 필터링 

##### ✓ 요구사항 
문장을 입력받아, **길이가 N 이상인 단어만 남기는 함수**를 작성하시오. 

1. 첫 줄: 문장 (공백으로 단어 구분)

2. 둘째 줄: 정수 N

3. 길이가 N 이상인 단어들을 **원래 순서대로** 한 줄에 공백으로 출력

```text
입력
python is very interesting
4

출력
python very interesting

````python
text = list(map(str, input("문장 입력: ").split()))
number = int(input("정수 입력: "))

text_pass = []

for word in text:
    if len(word) >= number:
        text_pass.append(word)
print(f"{' '.join(text_pass)}")
````

#### 문제 3. 완전수(perfect number) 찾기

##### ✓ 요구사항 

자연수 n이 **자기 자신을 제외한 약수의 합이 자신과 같으면** 완전수라고 한다 

예: 6의 약수 1 + 2 + 3 = 6 → 완전수

1. 정수 N을 입력받아 

    **1 이상 N 이하의 완전수들을 모두 출력**하는 프로그램을 작성하시오 

2. 없으면  `"No perfect number"` 출력

```text
입력: 30
출력: 6 28

````python
N = int(input("정수 입력: "))

result = []

for i in range (1, N + 1):
    
    div_sum = 0 

    for j in range(1, i):
        if i % j == 0:
            div_sum += j 

    if div_sum == i:
        result.append(div_sum)

if result:
    print(' '.join(str(x) for x in result))
else:
    print("No perfect number")
````

#### 문제 4. 안전한 정수 입력 + 1부터 N까지 합

##### ✓ 요구사항 

사용자에게 정수 N을 입력받아 **1부터 N까지의 합**을 구하는 데, 

정수를 잘못 입력하면 (예: 'abc') 다시 입력받도록 하는 프로그램을 작성하시오 

1. `safe_int_input(prompt)`함수 안에서 `try ~ except`를 사용하여 예외처리 

2. 올바른 정수가 들어올 때까지 무한 반복

```text
N 입력: abc
올바른 정수를 입력하세요!
N 입력: 10
1부터 10까지의 합은 55입니다. 

```python
# 숫자를 안전하게 입력받는 함수
def safe_int_input(msg):
    #올바른 숫자가 입력될 때까지 반복해서 입력받는 함수
    while True:                             # 올바른 입력이 들어올 때까지 반복
        try:
            return int(input(msg))          # 숫자로 변환 시도
        except ValueError:                  # 숫자가 아니면 안내 메시지 출력
            print("숫자가 아닙니다. 다시 입력해주세요.")

````python
def safe_int_input(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("올바른 정수를 입력하세요!")

N = safe_int_input("정수 입력: ")

sum = 0 

for i in range(1, N + 1):
    sum = sum + i 

print(f"1부터 {N}까지의 합은 {sum}입니다")
````

```python
# for 루프 대신 사용할 수 있는 코드 (O(1) 시간 복잡도)

print(f"1부터 {number}까지의 합은 {total_sum}입니다")

#### 문제 5. 파일에서 정수 읽어 평균 구하기 (예외처리 포함)

##### ✓ 요구사항 

1. 첫 줄에 **파일 이름**을 입력받는다 

    예: `numbers.txt`

2. 파일에는 한 줄에 하나씩 정수가 적혀 있다고 가정한다 

3. 파일을 읽어 오는 모든 정수의 **평균을 출력**한다 

4. 예외처리:

    - 파일이 없으면: `"파일을 찾을 수 없습니다."` 출력
    - 정수가 아닌 값이 섞여 있으면: `"정수로 변환할 수 없는 값이 없습니다."` 출력

````python
name = input("파일 이름: ")

try:
    with open(name, 'r') as file:
        
        # 파일 내용을 읽고 평균을 계산
        sum = 0 
        cnt = 0 

        for numbers in file:
            try:
                number = int(numbers.strip())
                sum += number       
                cnt += 1
            except ValueError:
                print("정수로 변환할 수 없는 값이 없습니다.")
                break

        if cnt > 0:
            print(sum / cnt)

except FileNotFoundError:
    print("파일을 찾을 수 없습니다.")
````

- strip():  문자열 앞뒤의 공백 또는 특별한 문자 삭제
- split():  문자열 내부에 있는 공백 또는 특별한 문자를 구분해서, 리스트로 변환

#### 문제 6. 간단 사칙연산 계산기 (예외처리 포함)

##### ✓ 요구사항 

다음 형태로 수식을 입력받아 계산 결과를 출력하는 프로그램을 작성하시오 

- 입력 예: `10 / 3`, `7 / 8`, `5 + 2`, `9 - 4`

- 지원 연산자: `+`, `-`, `*`, `/`

- 예외처리: 

    - 정수가 아닌 값 입력 → `"숫자를 올바르게 입력하세요."`

    - 0으로 나누기 → `"0으로 나눌 수 없습니다."`

    - 지원하지 않는 연산자 → `"지원하지 않은 연산자입니다."`

````python
math = input("수식 입력: ").split()

result = 0 

try:
    x = int(math[0])
    y = int(math[2])
    
    op = math[1]

    ops = {
        '+': lambda a, b: a + b,        # lambda를 쓰지 않으면, 연산자별로 if–elif–else 분기문을 사용
        '-': lambda a, b: a - b,
        '*': lambda a, b: a * b,
        '/': lambda a, b: a / b
    }

    if op in ops:
        result = ops[op](x, y)
    
        print(x, op, y)
        print(f"= {result}")

    else:
        print("지원하지 않은 연산자입니다.")

except ValueError:
    print("숫자를 올바르게 입력하세요.")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
````
