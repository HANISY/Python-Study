## 1번 문제 - 딕셔너리를 활용한 단어 빈도수 분석기

### ✓ 문제 설명

한 문장이 입력으로 주어진다.

문장 안의 모든 단어에 대해 **각 단어가 몇 번 등장했는지** 세고

이를 **사전순 오름차순으로 정렬하여 출력**하는 프로그램을 작성하시오

단어는 공백 단위로 구분한다

**입력 형식**

문자열 한줄

```python
apple banana apple orange banana apple

**출력 형식**
```python
apple 3
banana 2
orange 1
```

각 줄에 `"단어 빈도"` 형태로 출력한다

### ✓ 조건
- 대소문자 구문(Apple≠apple)
- 딕셔너리로 단어 빈도를 반드시 계산할 것
- 단어는 사전순 오름차순으로 정렬해서 출력

````python
words = input("문자열을 한줄로 입력하세요: ").split()

word_dict = {}

for word in words:
  if word in word_dict:
    word_dict[word] += 1
  else:
    word_dict[word] = 1

word_dict = dict(sorted(word_dict.items()))     # dict() 정렬된 튜플 리스트를 다시 딕셔너리 형태로 변환
                                                # sorted() 순회 가능한 객체(리스트, 튜플, 문자열 등)든 정렬하여 새로운 리스트를 반환하는 내장 함수

for key, value in word_dict.items():
  print(key, value)
````

## 2번 문제 - 튜플을 이용한 학생 성적 관리 프로그램

### ✓ 문제 설명

학생 이름과 정수가 `(이름, 정수)` 형태의 튜플로 주어진다.

학생 수 N이 입력되고, N개의 튜플이 들어온다.

이 중에서 평균 점수 이상을 받은 학생들의 이름을 사전순으로 출력하시오.

**입력 형식**

첫 줄: 정수 N
그 다음 N줄: `"이름 정수"` 형태

```python
4
Kim 80
Lee 70
Park 90
Choi 60

**출력 형식**

평균 점수 이상인 학생들의 이름을 사전순으로 출력한다

```python
Kim
Park

````python
num_student = int(input("학생 수:"))

student_list = []

for i in range(num_student):
  student = input("이름과 점수를 입력하세요(예 Lee 100):")
  student_list.append(tuple(student.split()))

# 학생 평균 계산
score_avg = sum(int(student[1]) for student in student_list) / num_student


# 평균 점수 이상인 학생들 이름 사전순으로 출력
name_high = [
    student[0] for student in student_list if int(student[1]) >= score_avg
]

for name in sorted(name_high):
  print(name)
````

## 3번 문제 - 스택을 이용한 "문자열 되돌리기"

### ✓ 문제 설명

문자열이 주어졌을 때, 스택(stack) 자료구조를 이용하여

**문자열을 뒤집어서 출력**하는 프로그램을 작성하시오.

<br>

단순히 `[::1]`같은 파이썬 기능을 쓰는 것이 아니라,

**스택 push/pop 동작을 직접 구현**해야 한다.

**입력 형식**

문자열 한줄

```python
HELLO

**출력 형식**

```python
OLLEH

````python
class Stack:
  def __init__(self):
    self.list = []

  def push(self, item):
    self.list.append(item)

  def pop(self):
    return self.list.pop()

text = input("문자열을 한줄로 입력하세요")

stack = Stack()
for char in text:
  stack.push(char)

while stack.list:
  print(stack.pop(), end='')    # 개행 없이 연속 출력
````
