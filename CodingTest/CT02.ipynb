##### 초급 문제 - 가장 많이 등장한 문자 찾기 

- **문제 설명** : <br>
영문 소문자로 이루어진 문자열 `s`가 주어진다. <br>
문자열에서 **가장 많이 등장한 문자와 그 빈도**를 출력하시오 <br>
가장 많이 등장한 문자가 여러 개라면 **사전순으로 가장 앞선 문자**를 출력한다

- **입력**: <br>
한 줄 문자열 `s` (1 ≤ len(s) ≤ 1,000)

- **출력**: <br>
`<문자> <빈도>` 형태로 출력

```python
text = input("문자열 입력> ")

count = {} 

for word in text:
    if word not in count: 
        count[word] = 0 
    count[word] += 1

max_char = ""
max_count = 0 

for key, value in count.items():
    if value > max_count:
        max_count = value
        max_char = key
    elif value == max_count:        # 가장 많이 등장한 문자가 여러 개라면 사전순으로 가장 앞선 문자 출력 
        if key < max_char:
            max_char = key

print("입력")
print(text)
print()
print("출력")
print(max_char, max_count)
```

##### 중급 문제 - 가장 긴 연속 문자 길이 

- **문제 설명** : <br>
문자열 `s`가 주어질 때, <br>
동일한 문자가 **연속으로 가장 길게 등장하는 구간의 길이**를 구하시오

- **입력**: <br>
한 줄 문자열 `s` (1 ≤ len(s) ≤ 1,000)

- **출력**: <br>
가장 긴 연속 구간의 길이 (정수)

```python
text = input("문자열 입력> ")

max_text = ''
max_count = 0

previous = text[0]      # 첫 글자를 ‘현재 연속 중인 문자’로 기준 설정
count = 1

for i in range(1,len(text)):        # 두 번째 글자(index 1)부터 비교 시작, 문자열 길이 기준으로 반복 
    current = text[i]

    if current == previous: 
        count += 1 
    else: 
        if count > max_count:
            max_count = count
            max_text = previous
        
        count = 1
        previous = current    


# 마지막 연속 구간을 비교해야함 
if count > max_count:
    max_count = count
    max_text = previous


print("입력")
print(text)
print()
print("출력")
print(max_count)
```

##### 고급 문제 - 부분 문자열 중 숫자만 추출해 가장 큰 값 찾기  

- **문제 설명** : <br>
영문자 + 숫자로 이루어진 문자열이 주어진다 <br>
문자열에서 **숫자로 이루어진 모든 연속 문자열**을 추출한 뒤, <br>
그 중 **가장 큰 숫자**를 정수로 출력하시오. 

```python
text = input("문자열 입력> ")

number = []
current = ""

for char in text:
    if char.isdigit():
        current += char
    else:
        number.append(current)
        current = ""

if char.isdigit():
    current += char
    number.append(current)

print(number)
```

```python
text = input("문자열 입력> ")

number = []
current = ""

for char in text:
    if char.isdigit():
        current += char
    else:
        if current != "":        
            number.append(current)
        current = ""

if current != "":
    number.append(current)

print(number)
```

| char | digit? | current | number              |
| ---- | ------ | ------- | ------------------- |
| '1'  | yes    | "1"     | []                  |
| '2'  | yes    | "12"    | []                  |
| 'a'  | no     | ""      | ["12"]              |
| 'b'  | no     | ""      | ["12"]              |
| '3'  | yes    | "3"     | ["12"]              |
| '4'  | yes    | "34"    | ["12"]              |
| 끝남   | —      | "34"    | ["12"] ← "34" 아직 없음 |

```python
text = input("문자열 입력> ")

numbers = []
current = ""

for char in text:
    if char.isdigit():
        current += char
    else:                                  # 숫자가 아닌 문자를 만난 순간
        if current != "":                  # 버퍼가 비어있지 않으면
            numbers.append(current)        # 지금까지 모은 숫자 덩어리를 리스트에 저장
        current = ""

# 마지막 문자가 숫자로 끝난 경우
# 반복문 안에서는 '숫자가 끊어질 때'만 추가하므로
# 마지막이 숫자라면 추가할 기회가 없어서 여기서 한 번 더 처리

if current != "":
    numbers.append(current)

max_number = 0

for number in numbers:
    int_num = int(number)         # 문자열로 이루어진 숫자 덩어리를 정수로 변환
    if int_num > max_number:      # else가 없으면 조건이 False일 때 “아무것도 하지 않는다”는 의미
        max_number = int_num

print("입력")
print(text)
print()
print("출력")
print(max_number)
```
