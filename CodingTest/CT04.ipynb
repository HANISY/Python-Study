#### 중급 - 홀수끼리만 더하기 

##### ✓ 문제 설명

정수들이 주어진다 

이 중에서 **홀수끼리만 서로 바로 이어져 있는 경우**, 

그 연속된 홀수들의 합을 구하고, 

그 중 **가장 큰 합**을 출력하시오 

입력 예: 
```python
3 5 2 7 9 4 3 3 3 

출력 예: 
```python
19

````python
numbers = list(map(int, input("정수 입력: ").split()))

odds_sum = 0
sum_list = []

for number in numbers:
    if number % 2 != 0:
        odds_sum += number
    else:
        sum_list.append(odds_sum)
        odds_sum  = 0 
        
sum_list.append(odds_sum)       # 루프가 끝난 후 남아있는 마지막 합계 추가 

max = 0 

for sum in sum_list:
    if sum > max:
        max = sum 

print(max)
````

#### 초급 - 방향 전환 로봇 (간단 시뮬레이션) 

##### ✓ 문제 설명

한 로봇이 (0, 0) 위치에 서 있고 처음에는 **북쪽(N)** 을 바라보고 있다 

명령 문자열이 주어졌을 때 로봇의 **최종 위치 (x, y)** 를 출력하시오 

- `F`: 바라보는 방향으로 한 칸 전진

- `L`: 왼쪽으로 90도 회전

- `R`: 오른쪽으로 90도 회전

입력 예: 
```python
FFRFFLFF

출력 예:
```python
2 3

````python
commands = input("경로 입력: ")

# DELTAS N:(0, 1), E(1, 0), S(0, -1), W(-1, 0)
DELTAS = [(0, 1), (1, 0), (0, -1), (-1, 0)]

x, y = 0, 0 
d_index = 0 


for command in commands:
    if command == 'F':
        dx, dy = DELTAS[d_index]
        x += dx
        y += dy
    elif command =='R':
        d_index = (d_index + 1) % 4
    elif command =='L':
        d_index = (d_index - 1) % 4

print(x, y)
````

#### 고급 - 문자열 압축 코딩 (삼성 SW형 시뮬레이션)

##### ✓ 문제 설명

문자열이 주어진다. 이를 다음 규칙에 따라 **압축 문자열**로 변환하시오

압축 규칙:

1. 같은 문자가 **연속으로 반복될 경우**,

    `"문자 + 반복횟수"` 형태로 변환

2. 단, **반복 횟수가 1이면 숫자는 생략**한다 

3. 연속하지 않으면 그대로 이어 붙인다

입력 예:
```python
AAABCCDDDDX

출력 예:
```python
A3BC2D4X

설명 

- A 3번 → `A3`
- B 1번 → `B`
- C 2번 → `C2`
- D 4번 → `D4`
- X 1번 → `X`

조건

- 함수 없이 반복문으로만 구현 가능

- 삼성 기출 스타일의 문자열 시뮬레이션 문제 

````python
text = input("문자열 입력: ")

result = ''

previous = text[0]
count = 1

for i in range(1, len(text)):
    current = text[i]

    if current == previous:
        count += 1
    else:
        if count == 1:
            result += previous
        else: 
            result += previous + str(count)

        previous = current
        count = 1

if count == 1:
    result += previous
else:
    result += previous + str(count)

print(result)
````
