````python
# random 모듈

import random
print("# random 모듈")

# random(): 0.0 <= x < 1.0 사이의 float을 리턴합니다. 
print("- random():", random.random())

# uniform(min, max): 지정한 범위 사이의 float을 리턴합니다 
print("- uniform(10, 20):", random.uniform(10, 20))

# randrange(): 지정한 범위의 int를 리턴합니다. 
# - randrange(max): 0부터 max 사이의 값을 리턴합니다. 
# - randrange(min, max): min부터 max 사이의 값을 리턴합니다. 
print("- randrange(10):", random.randrange(10))

# choice(list): 리스트 내부에 있는 요소를 랜덤하게 선택합니다. 
print("- choice([1, 2, 3, 4, 5]):", random.choice([1, 2, 3, 4, 5]))

# shuffle(list): 리스트의 요소들을 랜덤하게 섞습니다.
print("- shuffle([1, 2, 3, 4, 5]):", random.shuffle([1, 2, 3, 4, 5]))

# sample(list, k=<숫자>): 리스트의 요소 중에 k개를 뽑습니다.
print("- sample([1, 2, 3, 4, 5], k=2):", random.sample([1, 2, 3, 4, 5], k=2))
````

````python
# datetime 모듈
import datetime

# 현재 시각을 구하고 출력
print("# 현재 시각 출력하기")
now = datetime.datetime.now()
print(now.year, "년")
print(now.month, "월")
print(now.day, "일")
print(now.hour, "시")
print(now.minute, "분")
print(now.second, "초")
print()

# 시간 출력 방법
print("# 시간을 포맷에 맞춰 출력하기")
output_a = now.strftime("%Y.%m.%d %H:%M:%S")
output_b = "{}년 {}월 {}일 {}시 {}분 {}초".format(now.year,\
                                            now.month,\
                                                now.day,\
                                                    now.hour,\
                                                        now.minute,\
                                                            now.second)
output_c = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")        # 앞에 *을 붙이면 요소 하나하나가 매개변수로 지정
print(output_a)
print(output_b)
print(output_c)
print()
````

````python
# 시간 처리하기 
import datetime
now = datetime.datetime.now()

# 특정 시간 이후의 시간 구하기
print("# datetime.timedelta로 시간 더하기")
after = now + datetime.timedelta(\
    weeks=1,\
        days=1,\
            hours=1,\
                minutes=1,\
                    seconds=1)
print(after.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))
print()

# 특정 시간 요소 교체하기
print("# now.replace()로 1년 더하기")
output = now.replace(year=(now.year + 1))
print(output.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))
````

**b'...'** 바이너리 데이터를 의미

````python
# operator.itemgetter() 함수
books = [{
    "제목": "혼자 공부하는 파이썬",
    "가격": 18000
},{
    "제목": "혼자 공부하는 머신러닝 + 딥러닝",
    "가격": 26000
},{
    "제목": "혼자 공부하는 자바스크립트",
    "가격": 24000
}]

def 가격추출함수(book):
    return book["가격"]

min(books, key=가격추출함수)                    # 가격추출함수가 무엇인지 알기 위해 함수를 찾아 내용을 살펴봐야함 
min(books, key=lambda book: book["가격"])      # 다른 사람이 코드를 읽는 것이 어려울 수 있음 
````

````python
# 특정 요소를 추출
from operator import itemgetter

books = [{
    "제목": "혼자 공부하는 파이썬",
    "가격": 18000
},{
    "제목": "혼자 공부하는 머신러닝 + 딥러닝",
    "가격": 26000
},{
    "제목": "혼자 공부하는 자바스크립트",
    "가격": 24000
}]

print("# 가장 저렴한 책")
print(min(books, key=itemgetter("가격")))
print()

print("# 가장 비싼 책")
print(max(books, key=itemgetter("가격")))
````

**함수 데코레이터**: 함수를 인자로 받아 기능을 감싸거나 확장해 새로운 함수를 반환하는 구조

**클래스 데코레이터**: 클래스를 인자로 받아 초기화·속성·메서드 등을 변경하거나 확장해 새로운 클래스를 생성하는 구조

````python
# 함수 데코레이터의 기본
# 들여쓰기 주의하기
def test(function):
    def wrapper():
        print("인사가 시작되었습니다.")
        function()
        print("인사가 종료되었습니다.")
    return wrapper
    
@test
def hello():
    print("hello")
    
hello()
````

함수 데코레이터를 사용할 때 매개변수 등을 전달할 수 있어 반복되는 구문이 많아질 때 소스의 가독성도 높이고 매우 유용하게 사용 가능

````python
from functools import wraps

def test(function):
    @wraps(function)
    def wrapper(*arg, **kwargs):
        print("인사가 시작되었습니다.")
        function(*arg, **kwargs)
        print("인사가 종료되었습니다.")
    return wrapper

@test
def hello():
    print("hello")

hello()
````

**패키지**

여러 모듈을 하나의 디렉터리 구조로 묶어 체계적으로 관리할 수 있게 해주는 파이썬의 기능

복잡한 기능을 모듈 단위로 나누어 구조화하고 재사용성을 높일 때 사용

> 패키지 만들기 (1)



**\_\_init\_\_.py** 해당 폴더가 패키지임을 알려주고, 패키지와 관련된 초기화 처리를 하는 파일

내부의 \_\_all\_\_ 리스트에 지정된 모듈들이 `from <패키지 이름> import *`를 할 때 전부 읽어 들여짐

> 패키지 만들기(2)



**텍스트 데이터** 사람이 읽을 수 있는 문자 형태의 데이터 <br>
문자(글자)를 저장하기 위해 문자 집합과 인코딩 방식이 필요

**바이너리 데이터** 0과 1(바이트)로 구성된 기계가 이해하는 데이터 <br>
이미지, 영상, 실행파일 등이 모두 바이너리 형태로 저장

**인코딩(Encoding)** 문자를 컴퓨터가 처리할 수 있도록 문자 → 숫자(바이트)로 변환하는 과정 <br>
예: ASCII, UTF-8

**디코딩(Decoding)** 인코딩된 숫자(바이트)를 다시 문자로 변환하는 과정 <br>
예: UTF-8 바이트 → 한글 문자열

````python
from urllib import request

url = "https://www.hanbit.co.kr/images/common/logo_hanbit.png"

req = request.Request(
    url,
    headers={
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36"
    }
)

target = request.urlopen(req)   
output = target.read()          

print(output)            

file = open("output.png", "wb")
file.write(output)
file.close()
````
