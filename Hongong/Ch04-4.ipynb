#### 4-4. 문자열, 리스트, 딕셔너리와 관련된 기본 함수 

```python
example_list = ["요소A", "요소B", "요소C"]

print("# 단순 출력")
print(example_list)
print()

print("# enumerate() 함수를 적용해 출력합니다")
print(enumerate(example_list))
print()

print("# list() 함수로 강제 변환해 출력합니다")
print(list(enumerate(example_list)))
print()

print("# for 반복문과 enumerate() 함수 조합해서 사용하기")
for i, value in enumerate(example_list): # enumerate*() 함수를 사용하면 for와 in 사이에 반복 변수를 2개 넣을 수 있습니다 
    print("{}번째 요소는 {}입니다.".format(i, value))
```

```python
# 딕셔너리의 items()함수와 반복문 조합하기

example_dictionary = {
    "키A": "값A",
    "키B": "값B",
    "키C": "값C"
}

print("# 딕셔너리의 items() 함수")
print("items():", example_dictionary.items())
print()

print("# 딕셔너리의 items()함수와 반복문 조합하기")
for key, element in example_dictionary.items():
    print("dictionary[{}] = {}".format(key,element))
```

```python
# 1 ~ 100 사이에 있는 숫자 중 2진수로 변환했을 떄 0이 하나만 포함된 숫자를 찾고, 그 숫자들의 합을 구하기

output = [i for i in range(1, 100 + 1) # list comprehension
          if "{:b}".format(i).count("0") == 1 ]
# 1부터 100까지 반복하면서, "{:b}".format(i).count("0") == 1 조건을 만족하는 i만 리스트에 넣어라.

for i in output:
    print("{} : {}".format(i, "{:b}".format(i)))
print("합계: ",sum(output))
```

```python
# 도전 문제 01. 리스트에서 몇 가지 종류의 숫자가 사용되었는지 구하는 프로그램 구현
# 딕셔너리 사용 

numbers = [1, 2, 3, 4, 1, 2, 3, 1, 4, 1, 2, 3]
count_num = {}

for n in numbers:
    if n in count_num:
        count_num[n] += 1
    else:
        count_num[n] = 1

print("{}에서".format(numbers))
print("사용된 숫자의 종류는 {}개 입니다.".format(len(count_num)))
print("참고: ",count_num)
```

```python
# 정답
a = [1, 2, 3, 4, 1, 2, 3, 1, 4, 1, 2, 3]
counters = {}

for n in a:
    if n not in counters:
        counters[n] = 0
    counters[n] += 1

print(f"{a}에서")
print(f"사용된 숫자의 종류는 {len(counters)}개 입니다")
print(f"참고: {counters}")
```

```python
# 도전 문제 02. 염기 서열을 입력했을 때 각각의 염기가 몇 개 포함되어 있는지 세는 프로그램 구현 
# DNA는 A(아데닌), T(티민), G(구아닌), C(사이토신)으로 이루어져 있습니다 
# 반복문 사용
nucleos = input("염기 서열을 입력해주세요: ")
counter = {
    "a" : 0,
    "t" : 0,
    "g" : 0,
    "c" : 0
}

for nucleo in nucleos:
    counter[nucleo] += 1

for key in counter:
    print(f"{key}의 개수: {counter[key]}")
```

```python
# 도전 문제 03-1. 염기 서열을 입력했을 때 어떤 코돈이 존재하는지 출력하는 프로그램 구현 
# 코돈 : 염기 3개가 묶여 있는 것 
# 3개씩 나누고 남는 염기 서열은 결과에서 무시합니다. 

nucleos = input("염기 서열을 입력해주세요: ")

for i in range(0, len(nucleos), 3):
    codon = nucleos[i:i + 3]
    if len(codon) == 3:
        print(codon)
```

```python
# 도전 문제 03-2. 염기 서열을 입력했을 때 어떤 코돈이 몇 개 존재하는지 출력하는 프로그램 구현 

nucleos = input("염기 서열을 입력해주세요: ")
counter = {}

for i in range(0, len(nucleos), 3):
    codon = nucleos[i:i + 3]
    if len(codon) == 3:
        if codon not in counter:
            counter[codon] = 0
        counter[codon] += 1

print(counter)
```

```python
# 도전 문제 04. 2차원 리스트 평탄화
data = [1, 2, [3, 4], 5, [6, 7], [8, 9]]

flat = []

for items in data:
    if type(items) == list:
        for item in items:
            flat.append(item)
    else:
        flat.append(items)

print(flat)
```
