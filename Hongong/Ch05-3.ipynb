- **튜플** : 함수와 함께 많이 사용되는 리스트와 비슷한 자료형<br>
한번 결정된 요소는 바꿀 수 없음

    (데이터, 데이터, 데이터, ...)

````python
# 요소를 하나만 가지는 리스트 생성
[273]

# 요소를 하나만 가지는 튜플 생성
(273, )
````

````python
# 리스트와 튜플의 특이한 사용
[a, b] = [10, 20]
(c, d) = (10, 20)

print("a=", a)
print("b=", b)
print("c=", c)
print("d=", d)
````

````python
# 괄호가 없는 튜플
tuple_test = 10, 20, 30, 40
print("# 괄호가 없는 튜플의 값과 자료형 출력")
print("tuple_test:", tuple_test)
print("type(tuple_test):", type(tuple_test))
print()

# 괄호가 없는 튜플 활용
a, b, c = 10, 20, 30
print("# 괄호가 없는 튜플을 활용한 할당")
print("a:", a)
print("b:", b)
print("c:", c)
````

````python
# 변수의 값을 교환하는 튜플
a, b = 10, 20

print("# 교환 전 값")
print("a:", a)
print("b:", b)
print()

a, b = b, a

print("# 교환 후 값")
print("a:", a)
print("b:", b)
print()
````

````python
# 여러 개의 값 리턴
def test():
    return (10, 20)

a, b = test()

print("a:", a)
print("b:", b)
````

````python
for i, value in enumerate([1, 2, 3, 4, 5, 6]):      # i, value는 (i, value) 형태의 튜플에서 괄호를 제거한 것
    print("{}번째 요소는 {}입니다.".format(i, value))
````

````python
# 튜플을 리턴하는 함수의 예(1)
a, b = 97, 40

print(a // b)       # 몫
print(a % b)        # 나머지
````

````python
# 튜플을 리턴하는 함수의 예(2)
a, b = 97, 40
print(divmod(a, b))

x, y = divmod(a, b)
print(x)
print(y)
````

- **람다** : 1회용 함수<br>
매개변수로 함수를 전달하기 위해 함수 구문을 작성하는 것이 번거롭고, 코드 공간 낭비라는 생각이 들 때 함수를 간단하고 쉽게 선언하는 방법

    lambda 매개변수: 리턴값


- 콜백 함수 : 함수의 매개변수에 사용하는 함수

````python
# 함수의 매개변수로 함수 전달하기 
# 매개변수로 받은 함수를 5번 전달하는 함수 
def call_5_times(func):
    for i in range(5):
        func()

def print_hello():
    print("안녕하세요")

call_5_times(print_hello)       # 매개변수로 함수를 전달 
````

- **map() 함수** : 리스트의 요소를 함수에 넣고 리턴된 값으로 새로운 리스트를 구성

- **filter() 함수** : 리스트의 요소를 함수에 넣고 리턴된 값이 True인 것으로, 새로운 리스트를 구성

````python
def power(item):
    return item * item
def under_3(item):
    return item < 3

list_input_a = [1, 2, 3, 4, 5]

output_a = map(power, list_input_a)
print("# map() 함수의 실행 결과")
print("map(power, list_input_a):", output_a)        # 제너레이터 출력
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(under_3, list_input_a)
print("# filter() 함수의 실행 결과")
print("filter(under_3, list_input_a):", output_b)      
print("filter(under_3, list_input_a):", list(output_b))
````

````python
power = lambda x: x * x
under_3 = lambda x: x < 3

list_input_a = [1, 2, 3, 4, 5]

output_a = map(power, list_input_a)
print("# map() 함수의 실행 결과")
print("map(power, list_input_a):", output_a)       
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(under_3, list_input_a)
print("# filter() 함수의 실행 결과")
print("filter(under_3, list_input_a):", output_b)      
print("filter(under_3, list_input_a):", list(output_b))
````

````python
# 인라인 람다 

list_input_a = [1, 2, 3, 4, 5]

output_a = map(lambda x: x * x, list_input_a)        # 함수를 선언하지 않고 매개변수로 바로 입력 
print("# map() 함수의 실행 결과")
print("map(power, list_input_a):", output_a)        
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(lambda x: x < 3, list_input_a)
print("# filter() 함수의 실행 결과")
print("filter(under_3, list_input_a):", output_b)      
print("filter(under_3, list_input_a):", list(output_b))
````
