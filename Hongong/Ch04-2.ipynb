#### 4-2. 딕셔너리와 반복문

- 리스트 : 인덱스를 기반으로 값을 저장
- 딕셔너리 : 키를 기반으로 값을 저장 

```python
# 딕셔너리 선언
dictionary = {
    "name": "7D 건조 망고",
    "type": "당절임",
    "ingredient": ["망고", "설탕", "메타중아황산나트륨", "치자황색소"],
    "origin": "필리핀"
} # 선언할 때는 중괄호 사용 

# 출력
print("name:", dictionary["name"]) # 요소에 접근할 때는 대괄호 사용 
print("type:", dictionary["type"])
print("ingredient:", dictionary["ingredient"])
print("origin:", dictionary["origin"])
print()

# 값 변경하기 
dictionary["name"] = "8D 건조 망고"
print("name:", dictionary["name"])
```

```python
# 값 추가하기
dictionary["price"] = 5000
print(dictionary)
```

```python
# 기존 값 대체하기 
dictionary["name"] = "7D 건조 망고"
print(dictionary)
```

```python
# 값 삭제하기
del dictionary["ingredient"]
print(dictionary)
```

```python
# keyerror 발생
del dictionary["key"]
```

```python
# 키가 존재하는지 확인하고 값에 접근하기 
dictionary = {
    "name": "7D 건조 망고",
    "type": "당절임",
    "ingredient": ["망고", "설탕", "메타중아황산나트륨", "치자황색소"],
    "origin": "필리핀"
}

key = input("> 접근하고자 하는 키:")

if key in dictionary:
    print(dictionary[key])
else:
    print("존재하지 않는 키에 접근하고 있습니다")
```

```python
# 키가 존재하지 않을 때 None 출력 확인
dictionary = {
    "name": "7D 건조 망고",
    "type": "당절임",
    "ingredient": ["망고", "설탕", "메타중아황산나트륨", "치자황색소"],
    "origin": "필리핀"
}

# 존재하지 않는 키에 접근
# get()함수 사용
value = dictionary.get("존재하지 않는 키") # 키가 없을 때 에러를 발생시키지 않고 None을 반환
print("값:", value)

if value == None:
    print("존재하지 않는 키에 접근했었습니다")
```

```python
# for 반복문과 딕셔너리 
dictionary = {
    "name": "7D 건조 망고",
    "type": "당절임",
    "ingredient": ["망고", "설탕", "메타중아황산나트륨", "치자황색소"],
    "origin": "필리핀"
}

for key in dictionary:
    print(key, ":", dictionary[key])
```

```python
# 딕셔너리와 리스트
pets = [
    {"name":"구름", "age":5},
    {"name":"초코", "age":3},
    {"name":"아지", "age":1},
    {"name":"호랑이", "age":1}
]

print("# 우리 동네 애완 동물들") # 숫자와 문자열 사이에 빈칸이 없도록 출력 
for pet in pets:
    print("{} {}살".format(pet["name"],pet["age"]))
```

```python
# 숫자 등장 횟수 출력
# 숫자는 무작위로 입력해도 상관 없음
numbers = [1, 2, 6, 8, 4, 3, 2, 1, 9, 5, 4, 9, 7, 2, 1, 3, 5, 4, 8, 9, 7, 2, 3]
counter = {}

for number in numbers:
    if number in counter:
        counter[number] += 1
    else:
        counter[number] = 1

print(counter)
```

```python
# 값의 자료형 확인
character = {
    "name": "기사",
    "level": 12,
    "items":{
        "sword": "불꽃의 검",
        "armor": "풀플레이트"
    },
    "skill": ["베기", "세게 베기", "아주 세게 베기"]
}

for key in character:
    if type(character[key]) is str:
        print(key,":", character[key])
    elif type(character[key]) is list:
        for item in character[key]:
            print(key,":", item)
    elif type(character[key]) is dict: # 변수 이름을 선언 
        for sub_key in character[key]:
            print(sub_key, ":", character[key][sub_key]) # 선언한 변수명과 print() 안의 변수명을 반드시 일치시켜야 함
```
